Conclusion:

The exploratory data analysis (EDA) and data wrangling process revealed several insights into the SDOT Collision Dataset. The dataset contains 248,396 entries with 50 columns, including various features related to traffic accidents in Seattle.

Through EDA and data wrangling, the following conclusions were drawn:

- The dataset is comprehensive, covering a wide range of accident-related features.

- The distribution of accident severity is skewed, with a majority of accidents being non-severe.

- Correlation analysis revealed relationships between certain features, such as the number of injuries and person count, and accident severity.

Important Features for Predicting Accident Severity:

- SEVERITYCODE and SEVERITYDESC: These columns are the target variables, describing the severity of the accident.

- INJURIES, SERIOUSINJURIES, and FATALITIES: These columns provide information about the number of injuries and fatalities, which are strong indicators of accident severity.

- COLLISIONTYPE: This column describes the type of collision (e.g., vehicle, pedestrian, bicycle), which could influence the severity of the accident.

- PERSONCOUNT, PEDCOUNT, PEDCYLCOUNT, and VEHCOUNT: These columns provide information about the number of people and vehicles involved in the accident, which could impact the severity of the accident.

These features were identified through a combination of descriptive statistics, data visualization, and correlation analysis. They are expected to be useful in building a predictive model for accident severity.
